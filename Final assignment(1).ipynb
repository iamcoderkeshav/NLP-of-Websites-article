Function to get website data in textual file :

```python
def extract_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    title = soup.find('head').get_text(strip=True) if soup.find('head') else 'No Title Found'
    article = soup.find('p')
    tags=soup.find_all(True)
    
    if article:
        article_text = article.get_text(separator='\n', strip=True)
    else:
        article_text = 'No Article Text Found'
        print(response)
    
    return title, article_text
```

Directory to save the text files

```python
output_dir = 'Extracted_Data'
os.makedirs(output_dir, exist_ok=True)
```

Function to save all data text in text file

```python
for index, row in input_df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']
    
    # Extract article content
    title, article_text = extract_data(url)
    
    # Save the extracted content to a text file
    output_file = os.path.join(output_dir, f'{url_id}.txt')
    with open(output_file, 'w', encoding='utf-8') as file:
        file.write(f'Title: {title}\n\n')
        file.write(f'Article Text:\n{article_text}\n')

    print(f'Extracted content for URL ID {url_id} and saved to {output_file}')

print('Extraction process completed.')
```

Function to perform textual analysis

```python
def analyze_text(text):
    sentences = sent_tokenize(text)
    words = word_tokenize(text.lower())
    word_count = len(words)
    sentence_count = len(sentences)
    syllable_count = sum(textstat.syllable_count(word) for word in words)
    complex_words_count = len([word for word in words if textstat.syllable_count(word) >= 3])
    personal_pronouns_count = len([word for word, pos in nltk.pos_tag(words) if pos == 'PRP'])

    positive_score = sum(1 for word in words if word in positive_words)
    negative_score = sum(1 for word in words if word in negative_words)
    polarity_score = (positive_score - negative_score) / ((positive_score + negative_score) + 0.000001)
    subjectivity_score = (positive_score + negative_score) / (word_count + 0.000001)

    avg_sentence_length = word_count / sentence_count if sentence_count else 0
    avg_number_of_words_per_sentence = word_count / sentence_count if sentence_count else 0
    percentage_of_complex_words = (complex_words_count / word_count) * 100 if word_count else 0
    fog_index = 0.4 * (avg_sentence_length + percentage_of_complex_words)
    syllables_per_word = syllable_count / word_count if word_count else 0
    avg_word_length = sum(len(word) for word in words) / word_count if word_count else 0

    return {
        'POSITIVE SCORE': positive_score,
        'NEGATIVE SCORE': negative_score,
        'POLARITY SCORE': polarity_score,
        'SUBJECTIVITY SCORE': subjectivity_score,
        'AVG SENTENCE LENGTH': avg_sentence_length,
        'PERCENTAGE OF COMPLEX WORDS': percentage_of_complex_words,
        'FOG INDEX': fog_index,
        'AVG NUMBER OF WORDS PER SENTENCE': avg_number_of_words_per_sentence,
        'COMPLEX WORD COUNT': complex_words_count,
        'WORD COUNT': word_count,
        'SYLLABLE PER WORD': syllables_per_word,
        'PERSONAL PRONOUNS': personal_pronouns_count,
        'AVG WORD LENGTH': avg_word_length
    }
```

```python
output_file = 'final_output.xlsx'
output_data = []

for index, row in input_df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']
    with open(f'{url_id}.txt', 'r') as file:
        article_text = file.read()
        
        
    analysis_results = analyze_text(article_text)
    
    # Prepare the output row
    output_row = {col: row.get(col, '') for col in input_df.columns}
    output_row.update(analysis_results)
    
    output_data.append(output_row)

df_output = pd.DataFrame(output_data, columns=output_columns)
df_output.to_excel(output_file, index=False)

print('Text analysis and extraction process completed.')
```
